#### Calculate the average star rating for each genre, but only include genres with at least 10 movies


#### Option 1: manually create a list of relevant genres, then filter using that list

```python
# Answer:
genre_count = movies.genre.value_counts()
genre_list = ['Drama','Comedy','Action','Crime','Biography','Adventure','Animation','Horror','Mystery']
movies1 = movies[movies['genre'].isin(genre_list)]
movies1.groupby('genre')['star_rating'].mean()
```

#### Option 2: automatically create a list of relevant genres by saving the value_counts and then filtering

```python
# Answer:
genre_count = movies.genre.value_counts()
genre_count = genre_count[genre_count.values >= 10].index.tolist() #genre_count is now a list containing filtered genre
movies = movies[movies.genre.apply(lambda x: True if x in genre_count else False)]
movies.groupby('genre')['star_rating'].mean()
```

```python
filtered = movies.groupby('genre').filter(lambda x: x['title'].count() >= 10)
filtered.groupby('genre')['star_rating'].mean()
```

#### Option 3: calculate the average star rating for all genres, then filter using a boolean Series

```python
# Answer:
movies.groupby('genre')['star_rating'].mean()
```

#### Option 4: aggregate by count and mean, then filter using the count

```python
# Answer:
grouped = movies.groupby('genre',as_index=False).agg({'star_rating':['count','mean']})
grouped[grouped.iloc[:,1].values>=10]
```
